```python
df=pd.read_csv('/content/drive/MyDrive/LH/complete data/cctv_area면적계싼.csv',encoding='cp949')
```

```python
df
```

```python
df = df[['gid_2','cctv_area']]
```

```python
df['감시취약지수'] = 1-df['cctv_area']/10000
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_ = MinMaxScaler()
scaler_.fit(df[['감시취약지수']])
scale = scaler_.transform(df[['감시취약지수']]) * 100
df2=pd.DataFrame(scale, columns=['감시취약지수(100)'])
```

```python
final = pd.concat([df, df2], axis = 1)
final = final[['gid_2','감시취약지수(100)']]
final.columns = ['gid','감시취약지수']
```

```python
final
```

```python
final.to_csv('/content/drive/MyDrive/LH/complete data/격자별_감시취약지수_2021.csv', index=False)
```
